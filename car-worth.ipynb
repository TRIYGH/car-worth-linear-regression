## R-score is better but not an 80....     (0.8)

```python
columns = ['Mileage', 'Cylinder', 'Liter', 'Doors', 'Cruise', 'Sound', 'Leather']
```

```python
r_scores = []
num_series = []
from itertools import combinations      #permutations
for i in range(len(columns)):
    for p in combinations(columns, i+1):
        q = list(p)
        x = car_data[q]
        y = car_data[['Price']]
        regr = linear_model.LinearRegression()
        result = regr.fit(x,y)
        r_scores.append(regr.score(x,y))
        num_series.append(i+1)
#         print(r_scores)

print(max(r_scores))
```

###   the MAX is best if using all the columns

#### -------------  the next 2 plots are just to prove that it's highly likely that using ALL columns will yield the best R-score  ------------

```python
plt.plot(r_scores)
```

```python
plt.scatter(r_scores,num_series)
```
